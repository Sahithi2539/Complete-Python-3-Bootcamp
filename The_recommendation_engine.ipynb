```python
#most_viewed, filters
import pandas as pd
import numpy as np

from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
all_gen = []
for i in df['genre']:
    for gen in i.split(','):
        if gen.replace('[','').replace("'","").replace(']','') not in all_gen:
            all_gen.append(gen.replace('[','').replace("'","").replace(']',''))
```

```python
df['genre'][0].split(',')[4].replace('[','').replace("'","").replace(']','')
```

```python
new_gen = []
for i in all_gen:
    new_gen.append(i.strip())
```

```python
all_gen = list(set(new_gen))
```

```python
for i in all_gen:
    df[i] = ''
```

```python
k = 0
for col in df['genre']:
    for gen in all_gen:
        if gen in col:
            df[gen][k] = 1
        else:
            df[gen][k] = 0
    k +=1
```
